```python
#나눔고딕 설치(설치 후 런타임 재시작하기)
#!sudo apt-get install -y fonts-nanum
#!sudo fc-cache -fv
#!rm ~/.cache/matplotlib -rf

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


plt.rc('font', family='NanumBarunGothic')


f=open('/content/BTC_USDCoinDesk.csv', 'r')
coindesk_data=pd.read_csv(f,header=0)
seq=coindesk_data[['Closing Price (USD)']].to_numpy()
print('데이터 길이:', len(seq), '/n앞쪽 10개 값:', seq[0:10])
plt.plot(seq,color='red')
plt.title('비트코인 가격')
plt.xlabel('Days'); plt.ylabel('Closing Price (USD)')
plt.show()

mean_value = np.mean(seq)
median_value = np.median(seq)
max_value = np.max(seq)
min_value = np.min(seq)

plt.figure(figsize=(16, 10))

# 1. 평균값 (그래프)
plt.subplot(3, 2, 1)
plt.plot(seq, label='Closing Prices', color='blue', alpha=0.7)
plt.axhline(mean_value, color='red', linestyle='--', label=f'Mean: {mean_value:.2f}')
plt.title('비트코인 평균값')
plt.xlabel('Days')
plt.ylabel('Closing Price (USD)')
plt.legend()

# 2. 중간값 (바형 그래프)
plt.subplot(3, 2, 2)
seq_1d = seq.flatten()
days = range(len(seq_1d))
plt.bar(days, seq_1d, color='blue', alpha=0.7, label='Closing Prices')
plt.axhline(median_value, color='green', linestyle='--', label=f'Median: {median_value:.2f}')
plt.title('비트코인 중간값 (바형 그래프)')
plt.xlabel('Days')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.tight_layout()
plt.show()

# 3. 최대값 (면적 그래프)
plt.subplot(3, 2, 3)
seq_1d = seq.flatten()
days = range(len(seq_1d))
plt.fill_between(days, seq_1d, color='blue', alpha=0.3, label='Closing Prices Area')
plt.plot(days, seq_1d, color='blue', alpha=0.7, label='Closing Prices')  # 라인 그래프 추가
plt.axhline(max_value, color='purple', linestyle='--', label=f'Max: {max_value:.2f}')
plt.title('비트코인 최대값 (면적 그래프)')
plt.xlabel('Days')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.tight_layout()
plt.show()

# 4. 최소값 (히스토그램)
plt.subplot(3, 2, 4)
seq_1d = seq.flatten()
plt.hist(seq_1d, bins=60, color='blue', alpha=0.7, label='Closing Prices')
plt.axvline(min_value, color='orange', linestyle='--', label=f'Min: {min_value:.2f}')
plt.title('비트코인 최소값 (히스토그램)')
plt.xlabel('Closing Price (USD)')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()

plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 6))
plt.plot(seq, color='blue', alpha=0.7, label='Closing Prices')
plt.axhline(mean_value, color='red', linestyle='--', label=f'Mean: {mean_value:.2f}')
plt.axhline(median_value, color='green', linestyle='--', label=f'Median: {median_value:.2f}')
plt.axhline(max_value, color='purple', linestyle='--', label=f'Max: {max_value:.2f}')
plt.axhline(min_value, color='orange', linestyle='--', label=f'Min: {min_value:.2f}')
plt.title('52주 비트코인 가격 및 통계')
plt.xlabel('Days')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.show()


coindesk_data['5-day MA'] = coindesk_data['Closing Price (USD)'].rolling(window=5).mean()
coindesk_data['20-day MA'] = coindesk_data['Closing Price (USD)'].rolling(window=20).mean()
coindesk_data['60-day MA'] = coindesk_data['Closing Price (USD)'].rolling(window=60).mean()
plt.figure(figsize=(12, 6))
plt.plot(coindesk_data['Closing Price (USD)'], label='Closing Prices', color='blue', alpha=0.7)
plt.plot(coindesk_data['5-day MA'], label='5-day MA', color='red', linestyle='--', alpha=0.9)
plt.plot(coindesk_data['20-day MA'], label='20-day MA', color='green', linestyle='--', alpha=0.9)
plt.plot(coindesk_data['60-day MA'], label='60-day MA', color='purple', linestyle='--', alpha=0.9)
plt.title('비트코인 가격과 이동 평균선')
plt.xlabel('Days')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.tight_layout()
plt.show()

```
